## Converting data types

### orders dataset

```python
orders.info(memory_usage='deep')
```

```python
date_columns = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

for col in date_columns:
    orders[col] = pd.to_datetime(orders[col])
```

```python
orders.info(memory_usage='deep')
```

### reviews dataset

```python
reviews.info(memory_usage='deep')
```

```python
review_dates = [
    'review_creation_date',
    'review_answer_timestamp'
]

for col in review_dates:
    reviews[col] = pd.to_datetime(reviews[col])
```

```python
reviews.info(memory_usage='deep')
```

## Data Cleaning Summary

## Conclusion

The data cleaning process focused on preserving data integrity while preparing the datasets for analysis. Missing values with valid business meaning, such as delivery timestamps for cancelled orders and optional review comments, were retained. Exact duplicate records in the geolocation dataset were removed because they did not provide additional information. Product records with missing category information were preserved to avoid unnecessary data loss and will be excluded only in category-specific analyses. This approach ensures that the original business data remains intact while maintaining analytical reliability.
